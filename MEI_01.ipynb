```python
# Importing Necessary packages
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import time
```

```python
# Constant values
GOAL_STATE = 100
INITIAL_STATE = 1
MAX_STEPS = 1000
EPISODES = 30
```

```python
# creating state transition

def state_transition(state, action):
    if action=="up":
        if state > 10:
            return state -10
    elif action=="down":
        if state < 91:
            return state +10
    elif action=="right":
        if state % 10 !=0:
            return state +1
    elif action=="left":
        if state % 10 != 1:
            return state - 1
    return state 
            
```

```python
# select random action of the smiley 

def action_robo():
    return random.choice(["right","left", "up", "down"])
```

```python
# creating a reward function

def reward_function(state):
    return 100 if state==GOAL_STATE else 0
```

```python
# Creating the function for end of episode

def end_episode(steps, state):
    return steps >= MAX_STEPS or state == GOAL_STATE
    
```

```python
# Creating the simulation of Robot

def robo_simulation():
    total_rewards = 0
    steps_to_goal = 0
    total_scores=[]
    state = INITIAL_STATE
    while not end_episode (steps_to_goal, state):
        action = action_robo()
        next_state = state_transition(state, action)
        scores = reward_function(next_state)
        total_scores.append(scores)
        total_rewards += scores
        steps_to_goal +=1
        if steps_to_goal >= MAX_STEPS or state == GOAL_STATE:
            next_state = INITIAL_STATE
        else:
            state = next_state
        
    return total_rewards, steps_to_goal, total_scores
    
```

```python
# Run episodes

steps_to_goal_list = []
total_runtimes = []
rewards_per_step = []

for episode in range(1, EPISODES+1):
    start_time = time.time()
    total_rewards, steps_to_goal, total_scores = robo_simulation()
    end_time = time.time() 
    rewards_per_step.append(total_rewards / steps_to_goal)
    total_runtimes.append(end_time - start_time) 
    steps_to_goal_list.append(steps_to_goal)
    
    
```

```python
# Calculating the values

avg_steps = np.mean(steps_to_goal_list)
std_dev_steps = np.std(steps_to_goal_list)
avg_reward_per_step = np.mean(rewards_per_step)
std_dev_reward_per_step = np.std(rewards_per_step)
avg_runtime = np.mean(total_runtimes)
std_dev_runtime = np.std(total_runtimes)

# Print and display results
print(f"Average Steps to Reach Goal: {avg_steps:.2f} (std dev: {std_dev_steps:.2f})")
print(f"Average Reward per Step: {avg_reward_per_step:.2f} (std dev: {std_dev_reward_per_step:.2f})")
print(f"Average Runtime per Episode: {avg_runtime:.4f} seconds (std dev: {std_dev_runtime:.4f} seconds)")

# Plotting the calculations in boxplot 

plt.figure(figsize=(10, 6))

plt.subplot(131)
plt.boxplot(steps_to_goal_list,notch=True,  # notch shape
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     ) 
plt.title('Steps to Reach Goal')
plt.xlabel('Episodes')

plt.subplot(132)
plt.boxplot(rewards_per_step,notch=True,  # notch shape
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     ) 
plt.title('Reward per Step')
plt.xlabel('Episodes')

plt.subplot(133)
plt.boxplot(total_runtimes,notch=True,  # notch shape
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     ) 
plt.title('Episode Runtimes')
plt.xlabel('Episodes')

plt.show()
```
